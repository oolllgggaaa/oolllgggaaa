## Обучение

### Добавление признаков

Так как видна ежедневная сезонность - то предлагаю добавить признаки день, неделя и месяц, и отнесем их к категориальному типу. 

```python
#Добавим признаки
taxi['month'] = taxi.index.month
taxi['dayofweek'] = taxi.index.dayofweek
taxi['hour'] =taxi.index.hour

taxi['month'] = taxi['month'].astype('category')
taxi['dayofweek'] = taxi['dayofweek'].astype('category')
taxi['hour'] = taxi['hour'].astype('category')
taxi
```

Так как мы измеряем часами, зададим MAX_LAG = 24

```python
# Задаем значение MAX_LAG 
MAX_LAG = 24
```

Зададим значения за предыдущее время (LAG)

```python
for lag in range(1, MAX_LAG+1):
    taxi['lag_{}'.format(lag)] = taxi['num_orders'].shift(lag) 
taxi.dropna(inplace=True)
```

Зададим скользящее время


```python
taxi['rolling_mean'] = taxi['num_orders'].shift().rolling(1).mean()
taxi.dropna(inplace=True)
taxi.head()
```

### Разделим данные на обучающую и тестовую выборки

```python
features = taxi.drop(['num_orders'], axis=1)
target = taxi['num_orders']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, shuffle=False)
```

### Приступим к обучению

LinearRegression

```python
#Обучим LinearRegression и добавим модель и рассчитанный RMSE в массив
train_model = []
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
rmse = mean_squared_error(y_test, predictions, squared=False)
print("RMSE:", rmse)
train_model.append([model, rmse])
```

DecisionTreeRegressor

```python
model = DecisionTreeRegressor(random_state=12345)
parameters = {
    'max_depth': range(1, 11, 2),
    'min_samples_leaf': range(1, 11, 2)
}
tscv = TimeSeriesSplit(n_splits=2)
grid_search = GridSearchCV(estimator=model, cv=tscv, param_grid=parameters, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
print("Лучшие параметры DecisionTreeRegressor:", best_params)
predictions = grid_search.predict(X_train)
rmse = mean_squared_error(y_train, predictions, squared=False)
print("RMSE DecisionTreeRegressor на обучающей выборке:", rmse)
train_model.append([grid_search.best_estimator_, rmse])
```

LGBMRegressor

```python
#Обучим LGBMRegressor и добавим модель и рассчитанный RMSE в массив
model = lgb.LGBMRegressor()
model.fit(X_train, y_train, callbacks=[lgb.log_evaluation()])
predictions = model.predict(X_train)
rmse = mean_squared_error(y_train, predictions, squared=False)
print("RMSE на обучающей выборке:", rmse)
train_model.append([model, rmse])
```

CatBoost

```python
#Обучим CatBoost и добавим модель и рассчитанный RMSE в массив
cat_features = X_train.select_dtypes(include=['category']).columns.tolist()
model = CatBoostRegressor(verbose=False)
model.fit(X_train, y_train, cat_features=cat_features, verbose=False)
predictions = model.predict(X_train)
rmse = mean_squared_error(y_train, predictions, squared=False)
print("RMSE на обучающей выборке:", rmse)
train_model.append([model, rmse])
```

```python
# Вывод результатов всех моделей
print("Результаты всех моделей:")
for i, (model, rmse) in enumerate(train_model):
    print(f"Модель {i + 1}: {type(model).__name__}, RMSE: {rmse}")
# Выбор лучшей модели
best_model, best_rmse = min(train_model, key=lambda x: x[1])  # Выбираем модель с минимальным RMSE
print("\nЛучшая модель:")
print(f"Тип модели: {type(best_model).__name__}, RMSE: {best_rmse}")
```

### Вывод по разделу

 На данном этапе мы добавили признаки день, неделя и месяц, и отнесли их к категориальному типу. Задали LAG и скользящее время.
На основе проведенного анализа и сравнения моделей, лучшей моделью для прогнозирования количества заказов такси является LGBMRegressor (LightGBM), которая показала наименьшее значение RMSE — 11.71. Это означает, что данная модель наиболее точно предсказывает количество заказов по сравнению с другими моделями.

## Тестирование

```python
# Извлекаем модели из списка train_models
model_lr = train_model[0][0]  
model_gs = train_model[1][0]  
model_lgb = train_model[2][0]  
model_cat = train_model[3][0]  
# Предсказания на тестовой выборке
# LinearRegression
preds_lr_test = model_lr.predict(X_test)
# GridSearchCV
preds_gs_test = model_gs.predict(X_test)
# LGBMRegressor
preds_lgb_test = model_lgb.predict(X_test)
# CatBoostRegressor
preds_cat_test = model_cat.predict(X_test)
# Вычисление RMSE на тестовой выборке
rmse_lr_test = mean_squared_error(y_test, preds_lr_test, squared=False)
rmse_gs_test = mean_squared_error(y_test, preds_gs_test, squared=False)
rmse_lgb_test = mean_squared_error(y_test, preds_lgb_test, squared=False)
rmse_cat_test = mean_squared_error(y_test, preds_cat_test, squared=False)
# Вывод результатов
print("RMSE LinearRegression на тестовой выборке:", rmse_lr_test)
print("RMSE GridSearchCV на тестовой выборке:", rmse_gs_test)
print("RMSE LGBMRegressor на тестовой выборке:", rmse_lgb_test)
print("RMSE CatBoostRegressor на тестовой выборке:", rmse_cat_test)

```

```python
# Выбор лучшей модели
results = [
    (model_lr, rmse_lr_test),
    (model_gs, rmse_gs_test),
    (model_lgb, rmse_lgb_test),
    (model_cat, rmse_cat_test)]
best_model, best_rmse = min(results, key=lambda x: x[1]) 
print("\nЛучшая модель:")
print(f"Тип модели: {type(best_model).__name__}, RMSE: {best_rmse}")
```

В будние дни спрос был выше, чем в выходные, что указывает на влияние рабочего графика на поведение клиентов. Кроме того, был обнаружен тренд роста заказов с марта по август, что может быть связано с сезонными или экономическими факторами.

Для построения модели были добавлены признаки, такие как день, неделя и месяц, а также созданы лаги и скользящие окна для учета временных зависимостей. В процессе обучения были протестированы различные модели, включая линейную регрессию, случайный лес и градиентный бустинг. Наилучшие результаты показала модель LGBMRegressor (LightGBM), которая достигла значения RMSE 38.9014 на тестовой выборке, что значительно ниже целевого порога в 48. Модель CatBoostRegressor также продемонстрировала высокую точность с RMSE 39.4588. Таким образом, LGBMRegressor была выбрана как наиболее эффективная для прогнозирования спроса на такси.

Итоговый вывод: модель LGBMRegressor успешно справляется с задачей прогнозирования количества заказов такси на следующий час, что позволяет компании оптимизировать распределение водителей в периоды пиковой нагрузки. Для дальнейшего улучшения модели можно рассмотреть использование более сложных алгоритмов, таких как LSTM, или добавление дополнительных признаков, например, погодных условий или информации о праздничных днях.
