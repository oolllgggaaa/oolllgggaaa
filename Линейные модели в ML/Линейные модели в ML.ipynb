## Описане проекта

Вы работаете в IT-компании, которая выполняет на заказ проекты по машинному обучению. К вам обратился фермер, владелец молочного хозяйства «Вольный луг». Он хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже. А продавцы и технологи так и норовят приукрасить своих коровок!
Поэтому он просит вас разработать модель машинного обучения, которая поможет ему управлять рисками и принимать объективное решение о покупке. «ЭкоФерма» готова предоставить подробные данные о своих коровах. Вам нужно создать две прогнозные модели для отбора бурёнок в поголовье:
Первая будет прогнозировать возможный удой коровы (целевой признак Удой);
Вторая — рассчитывать вероятность получить вкусное молоко от коровы (целевой признак Вкус молока).
С помощью модели нужно отобрать коров по двум критериям:
средний удой за год — не менее 6000 килограммов;
молоко должно быть вкусным.

## Описание данных

Файл ferma_main.csv содержит данные о стаде фермера на текущий момент. 

Описание данных:

#### 1.1 Проверим данные на наличие пропусков и дубликатов, корректны ли типы данных у каждого признака. 


```python
#Проверим пропуски и дубликаты
print(ferma_main.duplicated().sum())
print(ferma_dad.duplicated().sum())
print(cow_buy.duplicated().sum())
```

В датасете cow_buy слишком мало данных, чтобы удалять дубликаты, поэтому оставим их, а вот в ferma_main, предлагаю удалить их.

```python
#Удалим дубликаты
ferma_main = ferma_main.drop_duplicates()
print(ferma_main.duplicated().sum())
```

```python
#Приведем названия столбцов к нижнему регистру
ferma_main.columns  = ferma_main.columns.str.lower()
#Приведем названия столбцов к нормальному виду
ferma_main.rename(columns={'удой, кг': 'удой_кг',
                                        'эке (энергетическая кормовая единица)': 'эке',
                                        'сырой протеин, г' : 'сырой протеин_г',
                                        'спо (сахаро-протеиновое соотношение)' : 'спо',
                                        'порода папы_быка' : 'порода_папы',
                                        'тип пастбища' : 'тип_пастбища',
                                        'жирность,%' : 'жирность_%', 
                                        'белок,%' : 'белок_%', 'вкус молока' : 'вкус_молока'}, inplace=True)
# В столбце тип_пастбища орфографическая ошибка, исправим
ferma_main['тип_пастбища'].replace({'Равнинные': 'Равнинное'}, inplace=True)
# В столбце порода_папы орфографическая ошибка, исправим
ferma_main['порода_папы'].replace({'Айдиалл': 'Айдиал'}, inplace=True)
ferma_main.head()
```

```python
#Приведем названия столбцов к нижнему регистру
ferma_dad.columns  = ferma_dad.columns.str.lower()
#Приведем названия столбцов к нормальному виду
ferma_dad.rename(columns={'имя папы' : 'имя_папы'}, inplace=True)
ferma_dad.head()
```

```python
#Приведем названия столбцов к нижнему регистру
cow_buy.columns  = cow_buy.columns.str.lower()
#Приведем названия столбцов к нормальному виду
cow_buy.rename(columns={'тип пастбища': 'тип_пастбища',
                                        'порода папы_быка': 'порода_папы',
                                        'текущая_жирность,%' : 'жирность_%',
                                        'текущий_уровень_белок,%' : 'белок_%'}, inplace=True)
cow_buy['тип_пастбища'] = cow_buy['тип_пастбища'].str.capitalize()
cow_buy.head()
```

## 5.  Обучение модели линейной регрессии

### 5.1 Подготовим данные датасета ferma_main.csv и обучим на них первую модель LinearRegression.


#### 5.1.1 Используем в качестве целевого признак Удой, кг.


```python
# Используйте в качестве целевого признак Удой, кг.
X = ferma_main.drop(['удой_кг', 'белок_%', 'жирность_%', 'вкус_молока', 'id'], axis = 1)
y = ferma_main['удой_кг']
```

#### 5.1.2 Отберем признаки, которые подходят для модели линейной регрессии.


Предлагаю в категориальные признаки взять все категориальные столбцы, так как мы не знаем точную зависимость друг от друга.
А из количественных признаков, предлагаю исключить процент жирности и белка, в связи с полученными выше выводами зависимостей.

```python
# Разделяем на категориальные и количественные признаки 
cat_col_names = ['порода', 'тип_пастбища', 'порода_папы', 'возраст']
num_col_names = ['эке', 'сырой протеин_г', 'спо']
```

#### 5.1.3 Разбем данные на тренировочную и тестовую выборки.

```python
# Разбиваем данные на тренировочную и тестовую выборки
RANDOM_STATE = 42
X = ferma_main.drop('удой_кг', axis=1)
y = ferma_main['удой_кг']

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE)
```

#### 5.1.4 Кодируем категориальные признаки с помощью OneHotEncoder.


```python
# Кодируем категориальные признаки
encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names()
print("Закодированные признаки:", encoder_col_names)
```

#### 5.1.5 Масштабируем количественные признаки с помощью StandardScaler.

```python
# Масштабируем количественные признаки
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])
```

#### 5.1.6 Обучаем модель линейной регрессии на подготовленных данных.

```python
#Объединяем все преобразованные данные в тренировочный и тестовый датафреймы:
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
#Обучаем модель
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
predictions = model_lr.predict(X_test)
```

#### 5.1.7 Оценим качество модели на тестовой выборке: рассчитаем R², проанализируем остатки с помощью графиков.

```python
#Рассчитаем коэффициент детерминации
r2 = r2_score(y_test, predictions)
print(r2)
```

```python
# построим графики для анализа остатков
residuals = y_test - predictions
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(residuals, bins = 50)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(predictions, residuals)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

R2 = 0.78. При анализе дисперсии ошибок разных прогнозов можно заметить, что модель чаще и сильнее ошибается при предсказании удоя, когда он значительно выше или ниже среднего значения.
```python
# Используйте в качестве целевого признак Удой, кг.
X = ferma_main.drop(['удой_кг', 'белок_%', 'жирность_%', 'вкус_молока', 'эке','id'], axis = 1)
y = ferma_main['удой_кг']
# Разделяем на категориальные и количественные признаки 
cat_col_names = ['спо','порода', 'тип_пастбища', 'порода_папы', 'возраст']
num_col_names = ['сырой протеин_г', 'эке2']
# Разбиваем данные на тренировочную и тестовую выборки
RANDOM_STATE = 42
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    random_state=RANDOM_STATE) 
# Кодируем категориальные признаки
encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()
# Масштабируем количественные признаки
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])
#Объединяем все преобразованные данные в тренировочный и тестовый датафреймы:
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
#Обучаем модель
model_lr2  = LinearRegression()
model_lr2.fit(X_train, y_train)
predictions2 = model_lr2.predict(X_test)

#Рассчитаем коэффициент детерминации
r2 = r2_score(y_test, predictions2)
print(r2)
# построим графики по количественным признакам — гистограмму частотности распределения и диаграмму рассеяния
residuals2 = y_test - predictions2
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(residuals2, bins = 20)
axes[0].set_title('Гистограмма распределения остатков')
axes[0].set_xlabel('Остатки')

axes[1].scatter(predictions2, residuals2)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки')
axes[1].set_title('Анализ дисперсии')
plt.show()
```

## 6 Задача классификации

### 6.1 Подготовим данные датасета ferma_main.csv и обучим на них модель LogisticRegression.


```python
#Преобразуем признак вкуса молока для удобства
ferma_main.loc[ferma_main['вкус_молока'] == 'вкусно', 'вкус_молока'] = 1
ferma_main.loc[ferma_main['вкус_молока'] == 'не вкусно', 'вкус_молока'] = 0
ferma_main['вкус_молока'] = ferma_main['вкус_молока'].astype('bool')
```

```python
ferma_main.head()
```

```python
# Определяем названия столбцов
cat_col_names = ['спо', 'порода', 'тип_пастбища', 'порода_папы', 'возраст', 'имя_папы']
num_col_names = ['сырой протеин_г', 'эке2']
target_col_name = 'вкус_молока'

# Разделяем данные на признаки и целевую переменную
X = ferma_main[cat_col_names + num_col_names]
y = ferma_main[target_col_name]

# Создаем обучающую выборку (например, разбив данные на train и test)
X_train_ridge = X.copy()  # Можно заменить источником обучающего набора
ohe_features_ridge = cat_col_names  # Категориальные признаки для OHE
num_features = num_col_names  # На численные признаки

# Применяем OneHotEncoder
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)

# Удаляем незакодированные категориальные признаки (изначальные колонки)
X_train_ridge = X_train_ridge.drop(ohe_features_ridge, axis=1)  # Убедитесь, что ohe_features_ridge определены
X_encoded = encoder_ohe.fit_transform(X[ohe_features_ridge])
X_encoded = pd.DataFrame(X_encoded, columns=encoder_ohe.get_feature_names(ohe_features_ridge))

# Объединяем закодированные признаки и численные
X_train_ridge = pd.concat([X_train_ridge.reset_index(drop=True), X_encoded.reset_index(drop=True)], axis=1)

if 'спо_1.0' in X_train_ridge.columns:
    X_train_ridge = X_train_ridge.drop(columns=['спо_1.0'])

# Масштабирование численных признаков
scaler = StandardScaler()
X_train_ridge[num_features] = scaler.fit_transform(X[num_features])

# Смотрим на результат
print(X_train_ridge.head())

# Создаем и обучаем модель логистической регрессии
model_lr3 = LogisticRegression()
model_lr3.fit(X_train_ridge, y)

# Прогнозирование на тех же данных
y_pred = model_lr3.predict(X_train_ridge)
```

### 6.1.2 Оценим качество модели на тестовой выборке с помощью метрик accuracy, recall, precision.

```python
# Оценка качества модели
print(accuracy_score(y, y_pred))
print(recall_score(y, y_pred))
print(precision_score(y, y_pred))
```

### 6.1.2 Построим матрицу ошибок.

```python
# Расчет матрицы ошибок
conf_matrix = confusion_matrix(y, y_pred)

# Получаем уникальные метки классов
class_labels = y.unique()

# Отображаем матрицу ошибок
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=class_labels)
disp.plot(cmap='Blues')  # Выбор цветовой схемы
plt.title('Матрица ошибок')
plt.show()
```

### 6.2 Проанализируйте модель логистической регрессии.
*Объясните, какая ошибка критичнее всего при покупке коров: первого или второго рода. Помните, что главное для заказчика — вкусное молоко. Сформулируйте ответ с учётом метрик recall и precision, выберите более важную из них.*


Ошибки первого рода являются наиболее критичными при покупке коров, так как в этом случае мы можем рекомендовать фермеру коров с молоком, которое на самом деле не имеет хорошего вкуса.

Recall:
Эта метрика учитывает истинные положительные (TP) и ложные отрицательные (FN). Ошибки второго рода не так важны, поскольку нецелесообразно предлагать фермеру несколько коров с вкусным молоком. Куда серьезнее, если мы порекомендуем коров с невкусным молоком.

Precision:
Данная метрика основывается на истинных положительных (TP) и ложных положительных (FP) результатах. Поскольку количество FP (ошибок первого рода) нужно минимизировать, это становится критическим моментом. Следовательно, наша цель — максимизировать метрику Precision.

### 6.2.1 Сведем критичную ошибку к нулю

```python
# Выведем вероятности класса predict_proba() и посмотрим порог
porog = 0.5
while True:
    porog += 0.01
    y_pred = (model_lr3.predict_proba(X_train_ridge)[:,1] >= porog).astype(bool)
    if (precision_score(y, y_pred) == 1): break
porog
```

```python
y_pred = (model_lr3.predict_proba(X_train_ridge)[:,1] >= 0.82).astype(bool)
```

```python
cm = confusion_matrix(y, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.ylabel('True label')
plt.xlabel('Predicted');
```

Количество ошибок второго рода сильно увеличилось, но количество ошибок первого рода сведено к нулю, что нам и было необходимо.

```python
# Оценка качества модели
print(accuracy_score(y, y_pred))
print(recall_score(y, y_pred))
print(precision_score(y, y_pred))
```

**Предварительный вывод**

accuracy уменьшился с 0.62 до 0.46. recall сильно увеличился с 0.09 до 0.8, однако, нам необходимо было максимизировать precision, что мы и сделали, увеличив с 0.64 до 1, убрав вовсе ошибк ипервого рода. Это означает, что модель точно предскажет коров с вкусным молоком.
