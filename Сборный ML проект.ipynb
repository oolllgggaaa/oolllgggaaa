## Описание проекта

HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.

Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетвоён.

Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли.

### 1.2.2 Ошибки обработаны, теперь заменим пропуски с помощью пайплайна

```python
# Пайплайн для замены пропущенных значений
imputer_pipe = Pipeline(
    [
        (
            'SimpleImputer',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
    ]
)
```

```python
# Обработаем пропущенные значения при помощи пайплайна
nan_columns = ['dept', 'level', 'workload']
X_train[nan_columns] = imputer_pipe.fit_transform(X_train[nan_columns])
# Проверка
X_train.isnull().sum()
```

```python
# Разделяем данные на признаки и целевой признак
X_train1 = X_train.drop(['job_satisfaction_rate'], axis=1)
y_train1 = X_train['job_satisfaction_rate']
X_train1.shape
```

Если в тренировочной выборке мы разделяли данные, то в тестовой нужно их обьеденить

```python
# Разделяем данные на признаки и целевой признак
X_test1 = test.drop(['job_satisfaction_rate'], axis=1)
y_test1 = test['job_satisfaction_rate']
X_test1.shape
```

Поскольку целевой признак является непрерывным количественным, для обучения мы будем использовать регрессионные модели: линейную регрессию и дерево решений (для которого планируется перебор различных гиперпараметров).

```python
# создаём списки с названиями признаков
ohe_columns = ['last_year_violations', 'last_year_promo', 'dept']
ord_columns = ['level', 'workload']
num_columns = ['salary', 'employment_years', 'supervisor_evaluation']
```

```python
# Cоздаём пайплайны для подготовки признаков
ohe_pipe = Pipeline([
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
])

ord_pipe = Pipeline([
    ('ord', OrdinalEncoder(
        categories=[
            ['junior', 'middle', 'senior'],
            ['low', 'medium', 'high']
        ],
        handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
])

# Cоздаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
     ],
    remainder='passthrough'
)

# Cоздаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])
pipe_final
```

## 1.5 Обучение модели

Перед обучением модели, напишем функцию, которая принимает на вход массивы NumPy или объекты Series в pandas и возвращает значение метрики SMAPE. 

```python
#Создадим собственную метрику
def smape(y_true, y_pred):
    y_true = np.array(y_true)  # Преобразуем в numpy массив
    y_pred = np.array(y_pred)  # Преобразуем в numpy массив

    numerator = np.abs(y_pred - y_true)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    diff = np.divide(numerator, denominator, out=np.zeros_like(numerator), where=denominator != 0)

    return np.mean(diff) * 100

# Создаём пользовательскую метрику
smape_score = make_scorer(smape, greater_is_better=False)
```

Теперь, используем эту метрику при подборе гиперпараметров и оценке качества моделей. 

```python
# Создаем сетку гиперпараметров для множества моделей
param_grid = [[
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],

    [{
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 21),
        'models__min_samples_leaf': range(1, 6),
        'models__min_samples_split': range(2, 6),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],


    [{
        'models': [KNeighborsRegressor(n_jobs=-1, weights='distance')],
        'models__n_neighbors': range(5, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    }],
]

# Инициализируем список для хранения результатов
results = []

# В цикле переберем по 30 комбинаций гиперпараметров для каждой модели
for grid in param_grid:
    search = RandomizedSearchCV(
        pipe_final,
        grid,
        n_jobs=-1,
        cv=10,
        n_iter=30,
        scoring=smape_score,
        random_state=RANDOM_STATE
    )

    # Обучаем модель и сохраняем результаты
    search.fit(X_train1, y_train1)
    predictions_test = search.best_estimator_.predict(X_test1)

    results.append({
        'model': search.best_estimator_['models'],
        'train_SMAPE': (-1) * search.best_score_,
        'test_SMAPE': smape(y_test1, predictions_test)
    })

# Преобразуем результаты в DataFrame для удобного отображения
score_table = pd.DataFrame(results)
```

```python
# Выведем таблицу с результатами по рейтингу
scoring_table = pd.DataFrame(score_table)
scoring_table.sort_values(by='test_SMAPE')
```

**Предварительный вывод**

DecisionTreeRegressor показал наилучшие результаты по обеим метрикам, что делает его наиболее подходящим выбором из представленных моделей. Хорошие результаты по обоим метрикам, с train_SMAPE = 14.98 и test_SMAPE = 13.55, что предполагает хорошую обобщающую способность.



 Для обучения мы выбрали линейную регрессию и дерево решений с настройкой гиперпараметров, а также разработали собственную метрику SMAPE для оценки точности моделей. Этот подход обеспечил качественную подготовку данных и адекватную оценку регрессионных моделей.

```python
# Определим решетки параметров для настройки гиперпараметров различных моделей в процессе поиска наилучшей модели.
param_grid = [
    
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(4, 7),
        'models__max_features': range(6, 9),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },
 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },

    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 3),
        'preprocessor__num': [StandardScaler(), MinMaxScaler()],
    },

]
param_grid
```

```python
# Инициализируем списки для хранения результатов
models_end = []
train_score_end = []
test_score_end = []
best_score_end = 0
best_model_end = None  # Инициализируем переменную для хранения лучшей модели

# Инициализируем список для хранения результатов
results = []
max_iterations = min(30, len(param_grid))

# В цикле переберем по 30 комбинаций гиперпараметров для каждой модели и сохраним результаты в `score_table`
for grid in param_grid:
    search = RandomizedSearchCV(
    pipe_final_2,
    grid,
    n_jobs=-1,
    cv=10,
    n_iter=max_iterations,  # Устанавливаем n_iter
    scoring='roc_auc',
    random_state=RANDOM_STATE)

    # Обучаем модель
    search.fit(X_train2, y_train2)
    
    current_best_model = search.best_estimator_['models']
    predictions_test = search.best_estimator_.predict(X_test2)
    quit_predictions = search.best_estimator_.predict_proba(X_test2)[:, 1]
    current_train_score = search.best_score_
    current_test_score = roc_auc_score(y_test2, quit_predictions)

    print('Модель и её лучшие гиперпараметры:\n\n', current_best_model)
    print('ROC-AUC лучшей модели на кросс-валидации по обучающей выборке:', current_train_score)
    print(f'ROC-AUC на тестовой выборке: {current_test_score}')
    print('---')

    # Сохраняем результат итерации
    models_end.append(current_best_model)
    train_score_end.append(current_train_score)
    test_score_end.append(current_test_score)

    # Сохраняем данные для анализа важности
    if current_test_score > best_score_end:
        best_score_end = current_test_score
        best_model_end = current_best_model
        preprocess_params = search.best_estimator_.named_steps['preprocessor']  # Исправлено здесь
        
        # Получаем имена признаков
        feature_names_2 = preprocess_params.get_feature_names_out()  # Получение имен признаков

# Формирование финальной таблицы результатов
score_table_clf = {
    'model': models_end,
    'train_roc_auc': train_score_end,
    'test_roc_auc': test_score_end,
}

# Преобразуем результаты в DataFrame для удобного отображения
score_table_df_clf = pd.DataFrame(score_table_clf)

# Вывод таблицы результатов
print(score_table_df_clf)
```
